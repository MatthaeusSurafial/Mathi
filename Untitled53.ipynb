```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.preprocessing import StandardScaler

# Erstellen des DataFrames mit dem Datensatz YAZ
data = {
    'calamari': [6, 8, 6, 4, 7, 7, 3, 5, 5, 1, 2, 2, 8, 6],
    'fish': [6, 8, 5, 4, 4, 4, 7, 3, 13, 0, 3, 4, 2, 5],
    'shrimp': [12, 5, 11, 2, 9, 11, 3, 10, 10, 7, 5, 8, 7, 9],
    'chicken': [40, 44, 19, 28, 22, 33, 39, 35, 71, 32, 20, 17, 37, 45],
    'koefte': [23, 36, 12, 13, 18, 23, 27, 18, 46, 12, 18, 13, 21, 31],
    'lamb': [50, 37, 22, 28, 29, 23, 31, 38, 45, 29, 26, 17, 31, 44],
    'steak': [36, 30, 16, 22, 29, 37, 22, 37, 35, 18, 19, 17, 30, 27]
}

df = pd.DataFrame(data)

#zob die Menge von Chicken > 30 ist. Das habe ich gemacht um eine Zielvariable zu erstellen
df['target'] = (df['chicken'] > 30).astype(int)

# Merkmale und Zielvariable
X = df.drop('target', axis=1)
y = df['target']

# Überprüfen des DataFrames
df.head()
```

```python
#Trainieren des Deep Learning Modells und Generieren der Einbettungen

# Standardisierung um die merkmale numerisch zu machen
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# Eingabe für numerische Merkmale
numerical_inputs = Input(shape=(7,), name='numerical_inputs')
dense_layer = Dense(64, activation='relu')(numerical_inputs)
output_layer = Dense(1, activation='sigmoid')(dense_layer)
# Modell erstellen und zusammenführen
model = Model(inputs=numerical_inputs, outputs=output_layer)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
# Trainieren des Modells
model.fit(X_scaled, y, epochs=50, batch_size=4, validation_split=0.2)
#Einbettungen
embedding_model = Model(inputs=model.input, outputs=model.layers[-2].output)
# Generieren der Einbettungen
embeddings = embedding_model.predict(X_scaled)
```

```python
#Trainieren des XGBoost Modells mit den Einbettungen
import xgboost as xgb
from sklearn.metrics import accuracy_score

# XGBoost-Modell trainieren
xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_model.fit(embeddings, y)

# Vorhersagen und Bewertung
y_pred = xgb_model.predict(embeddings)
accuracy = accuracy_score(y, y_pred)
print(f'Accuracy: {accuracy:.4f}')
```

```python
# Standalone XGBoost-Modell trainieren
standalone_xgb_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
standalone_xgb_model.fit(X_scaled, y)
y_pred_standalone = standalone_xgb_model.predict(X_scaled)
accuracy_standalone = accuracy_score(y, y_pred_standalone)
print(f'Accuracy of Standalone XGBoost Model: {accuracy_standalone:.4f}')
```
